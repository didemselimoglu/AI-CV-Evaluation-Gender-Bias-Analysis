```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Lenovo\Desktop\Design_Applications\Resume dataset.csv")
```

```python
print(df['category'].value_counts())
```

```python
java_dev_df = df[df['category'] ==  'Java Developers/Architects Resumes'].copy()
java_resume_df = java_dev_df[java_dev_df['job_title'] == 'Java Developer'].copy()

java_resume_df.drop_duplicates()
```

```python
long_resumes = java_resume_df[java_resume_df['Text'].str.len() > 600]
final_resumes = long_resumes.drop_duplicates()
final_resumes
```

```python
selected_50 = final_resumes.sample(n=50, random_state=42).reset_index(drop=True)
selected_50
```

```python
selected_50.to_csv('final_java_resumes_50.csv', index=False)
```
